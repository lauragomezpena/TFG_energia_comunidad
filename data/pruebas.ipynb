```python
import pandas as pd
import numpy as np

ruta = "energiasedif1(11 pisosycomunes).csv"
df = pd.read_csv(ruta)
print(df.columns)
col_fecha = df.columns[0]
df[col_fecha] = pd.to_datetime(df[col_fecha], errors="coerce")
cols_sin_fecha = df.columns[1:]
mask_con_datos = df[cols_sin_fecha].notna().any(axis=1)

primera_fila_util = mask_con_datos.idxmax()
df_energia = df.loc[primera_fila_util:].reset_index(drop=True)

df_energia.head()
```

```python
df_energia.tail()
df_energia['POTENCIA ACTIVA'].max()
```

```python
print("Filas originales:", len(df))
print("Primera fila útil:", primera_fila_util)
print("Filas tras limpiar:", len(df_energia))
print("Fecha inicial útil:", df_energia.loc[0, col_fecha])
print("Fecha final:", df_energia.loc[len(df_energia)-1, col_fecha])
```
